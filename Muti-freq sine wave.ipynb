```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
%matplotlib notebook
fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-3, 3))
line, = ax.plot([], [], lw=2)
# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# The animation function is called sequentially
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * 3*(x - 0.01 * i))
    y1 = np.sin(2 * np.pi * (x - 0.01 * i))
    y5 = np.sin(2 * np.pi * 5 * (x - 0.01 * i))
    y = y + y1 + y5
    #y1 = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,


anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Animation for multi frequency sine wave")
anim.save('Mutlifreq_sine_wave.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
plt.show()
```
